```python
from gradientai import Gradient
def main():
  with Gradient() as gradient:
      base_model = gradient.get_base_model(base_model_slug="nous-hermes2")

      new_model_adapter = base_model.create_model_adapter(
          name="test model"
      )
      print(f"Created model adapter with id {new_model_adapter.id}")
      sample_query = "### Instruction: சரியான அளவை அடைந்த பொருள் மூலைகள் என்ன?"
      # \n\n### Response: சரியான அளவை அடைந்த பொருள் மூலைகள் விலையில் உள்ளன.
      print(f"Asking: {sample_query}")

      # before fine-tuning
      completion = new_model_adapter.complete(query=sample_query, max_generated_token_count=100).generated_output
      print(f"Generated (before fine-tune): {completion}")

      samples = [
        {'inputs': '### Instruction: நிர்வாகிகள் தற்காலிக பொருள் மூலைகளை எப்படி அளிக்கின்றன? \n\n### Response: நிர்வாகிகள் தற்காலிக பொருள் மூலைகளை அளிக்கும்போது பொருள் மூலைகள் சூழல் மாறாகும்.'},
        {'inputs': '### Instruction: நிர்வாகிகள் அந்தரங்க நிலைகளை எப்படி அளிக்கின்றன? \n\n### Response: நிர்வாகிகள் அந்தரங்க நிலைகளை பொருள் மூலைகளில் அளிக்கின்றன.'},
        {'inputs': '### Instruction: அந்தரங்க நிலைகளை அளித்து பொருள் மூலைகள் எப்படி அனுப்பப்படுகின்றன? \n\n### Response: அந\n\n்தரங்க நிலைகளை அளித்து பொருள் மூலைகள் சூழல் மாறாகும்போது செயல்படுகின்றன.'},
        {'inputs': '### Instruction: பொருள் மூலைகள் அளவை எப்படி கணக்கிடுகின்றன? \n\n### Response: பொருள் மூலைகள் அளவை மிகுதியான பொருள்களை அளவை எளிதாக கணக்கிடுகின்றன.'},
        {'inputs': '### Instruction: பொருள் மூலைகள் எத்தனை வகையாகப் பிரிக்கப்படுகின்றன? \n\n### Response: பொருள் மூலைகள் பல்வேறு வகையாகப் பிரிக்கப்படுகின்றன, அவை சரியான அளவை அடைந்து மிகுதியான பொருள்கள் மட்டுமே உள்ளன.'},
        {'inputs': '### Instruction: சூழல் மாறாகும்போது பொருள் மூலைகள் எப்படி செயல்படுகின்றன? \n\n### Response: சூழல் மாறாகும்போது, பொருள் மூலைகள் அந்தரங்க நிலைகளில் செயல்படுகின்றன.'},
        #{'inputs': '### Instruction: சரியான அளவை அடைந்த பொருள் மூலைகள் என்ன? \n\n### Response: சரியான அளவை அடைந்த பொருள் மூலைகள் விலையில் உள்ளன.'},
        {'inputs': '### Instruction: அதிகரிக்கும் விலைகள் பொருள் மூலைகள் என்ன? \n\n### Response: அதிகரிக்கும் விலைகள் பொருள் மூலைகள் இரண்டு முறை வேலை செய்கின்றன: ஒருவர் விலைகள் அதிகரிக்கும் போது, மற்றவர்கள் அதிகரிக்கும் விலைகள் குறைந்துவிடுகின்றன.'},
      ]

      # this is where fine-tuning happens
      # num_epochs is the number of times you fine-tune the model
      # more epochs tends to get better results, but you also run the risk of "overfitting"
      # play around with this number to find what works best for you
      num_epochs = 3
      count = 0
      while count < num_epochs:
          print(f"Fine-tuning the model, iteration {count + 1}")
          new_model_adapter.fine_tune(samples=samples)
          count = count + 1

      # after fine-tuning
      completion = new_model_adapter.complete(query=sample_query, max_generated_token_count=100).generated_output
      print(f"Generated (after fine-tune): {completion}")

      new_model_adapter.delete()

if __name__ == "__main__":
    main()
```
